```python
from pathlib import Path
import pandas as pd
import tarfile
import urllib.request

def load_housing_data():
    tarball_path = Path("datasets/housing.tgz")
    if not tarball_path.is_file():  # Check if the tarball file exists
        Path("datasets").mkdir(parents=True, exist_ok=True)  # Ensure the datasets directory exists
        url = "https://github.com/ageron/data/raw/main/housing.tgz"
        urllib.request.urlretrieve(url, tarball_path)  # Download the tarball
        with tarfile.open(tarball_path) as housing_tarball:
            housing_tarball.extractall(path="datasets")  # Extract the tarball

    # Check the directory for extracted files and adjust the path
    extracted_path = Path("datasets/housing.csv")
    if not extracted_path.is_file():  # If housing.csv is not directly in the datasets folder
        # Look inside a subfolder (check for housing directory)
        extracted_dir = Path("datasets/housing")  # Adjust according to the archive structure
        extracted_path = extracted_dir / "housing.csv"
        
    return pd.read_csv(extracted_path)  # Read and return the CSV file

# Load the housing data
housing = load_housing_data()
```
